# Model management

how software models is versionned?

**MAJOR**: incompatible API changes

**MINOR**: Adds functionnality in a backwards compatible manner

**PATCH**: makes backwards compatible bug fixes


model versioning proposal

**MAJOR**: Incompatibility in data or target variables

**MINOR**: model performnace is improved

**PIPELINE**: pipeline of model training is changed

**CI :Continous Integration**  ==> triggered when a new code is committed or pushed/ Build packages, container images / unit and integration tests / delivers the final packages to the continuous delivery pipeline.

*UNIT TESTING* testing each component in the pipeline to see if it is produced the expected artifact

==> **Unit test for data**:test for the feature engineered data

==> **Unit test for model**: test for functions/ test performance / test for subset



**CD: Continous Delivery** ==> deploys new and trained model to the target environment/ checks the prediction performance of the model before deploying.
