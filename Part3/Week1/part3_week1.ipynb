#------------- Neural Architecture search(NAS)-----------

* **Hyperparameter tuning**

-Types of parameters in ML Models:
  * Trainable params: learned by the algorithm

  * Hyperparams: Setted before launching the learning process/ not updated at each training step

        Hyperparameters are of two types:

        * Model hyperparameters which influence model selection such as the number and width of hidden layers

        * Algorithm hyperparameters which influence the speed and quality of the learning algorithm such as the learning rate for Stochastic Gradient Descent (SGD) and the number of nearest neighbors for a k Nearest Neighbors (KNN) classifier.

+Manual hyperparams tuning is not scalable even for ** shallow DNN ** [NN that only one hidden layer]

==> Keras Tuner : Tensorflow library to hypertune params automatically.



* **AutoML**

-Automates the entire ML workflow.

-It is not specific to a particular type of model

-NAS(Neural Architecture Serach) is a subfield of AutoML: a technique for automating the desgin of  ANN


    *  Search space:

    2 Types : MACRO & MICRO : They refer to different levels of granularity when exploring possible configurations for a machine learning model

       *MACRO:The macro search space refers to the larger, high-level decisions or configurations that define the architecture of a machine learning model or the hyperparameters of an algorithm. It includes fundamental choices such as the type and number of layers in a neural network, the size of each layer, the learning rate of an optimizer

       *Micro:focuses on the finer-grained choices within a specific configuration or architecture. It deals with choices like filter sizes in convolutional layers, the number of units in a recurrent cell, activation functions.
