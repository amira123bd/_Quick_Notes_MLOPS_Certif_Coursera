# Interpretability

- We should be able to query the model to understand :

+Why did the model behave in a certain way.

+how can we trust the predictions made by the model.

+what information can the model provide to avoid prediction errors.


* Interpretability methods

**1.Intrinsic or Post-Hoc**

+Intrinsic interpretability

applied when the model are intrinsically interpretable(linear and decision trees)

+Post-Hoc methods

treat models as black boxes

**2.Type of result**

+Feature summary stats

+Feature summary viz

+model internals

+Data points

**3.model specific / model agnostic**

+specific: designed for a specific model( for eg. intrinsically interpretable models are models specific)

+agnostic: applied to any model after it is trained / do not have access to the internals of a model

**4.Local / global**

explains an individual prediction or the global model behavior
