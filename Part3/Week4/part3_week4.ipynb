# Model Analysis

after developing, training and deploying a model it's time to seek for possible ways to improve the performance of our model:

**Evaluation**
* Black box evaluation( testing without knowing the internal architecture eg. **Tensorboard**)
 or

* model introspection( inspect model part by part)


**performance metrics vs optimization objectives**

* performance metrics: depends on the prob nature (like for eg. regression we can use RMSE and for classification accuracy, F1-score).

* Optimization objectives: loss function

**top level aggregate metrics vs slicing**

* Top level aggregate metrics: are calculated on the entire dataset

* slicing: to understand how the model is performing on a subset of the data



# TFMA: Tensorflow model analysis

TFMA is an open source, scalable framework for doing deep analysis of model performance, including analyzing performance on slices of data. TFMA is also a key part of a machine-learning pipelines, such as a TFX pipeline, to perform deep analysis before you deploy a newly trained version of your model.

# Continous integration and monitoring

-our training data is a snapshot of the world at a point of time
==> data change over time


**Data drift & shift**

concept drift: loss of prediction quality

concept Emergence: change in the distribution

types of dataset shift:

  +covariate shift: distribution of input data change/distribution of labels doesn't change

  +Prior probability shift: concept drift can be though as a type of proba shift


How to look for drift and shift?
* Supervised techniques = labels are needed (latency problems)

  2.Sequential analysis
