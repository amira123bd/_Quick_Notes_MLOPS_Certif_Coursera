* **Distributed Training**

With more data and larger models ==> models become more time-consuming.


==> we really need distributed model training techniques.


-Types of distributed training:

+Data parallelism: models are replicated into different accelerators(GPU/TPU) and data is split between them.

+Model parallelism: when models are too large and can't fit on a single device then they can be divided into partitions, assigning different partitions to different accelerators.
( workers need to synchronize the shared params ) ==> more complex

* **High-performnace ingestion**

High-performance ingestion is important in high-performance modeling.


Designing an efficient input pipeline is a requirement( using tf.data )


+How to optimize the pipeline performance:

 * Prefetching: begin loading data for the next step before the current step is complete

 * Parallelize data extraction and transformation(tf.data.AUTOTUNE to set paralleslism level automatically)

 * Caching
